```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target
feature_names = data.feature_names

# Perform PCA with 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
pca.components_.T
```

```python
pca.components_.shape
```

```python
# Create a biplot
def biplot(score, coeff, labels):
    plt.figure(figsize=(8, 8))
    plt.scatter(score[:, 0], score[:, 1], c=y, cmap='viridis')
    for i in range(len(coeff)):
        plt.arrow(0, 0, coeff[i, 0], coeff[i, 1], color='r', alpha=0.5)
        plt.text(coeff[i, 0] * 1.15, coeff[i, 1] * 1.15, labels[i], fontsize=12)

    plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]:.2%})")
    plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]:.2%})")
    plt.grid()

biplot(X_pca, pca.components_.T, feature_names)
plt.show()
```

```python
def biplot_with_numbers(score, coeff, labels):
    plt.figure(figsize=(8, 8))
    plt.scatter(score[:, 0], score[:, 1], c=y, cmap='viridis')
    for i in range(len(coeff)):
        plt.arrow(0, 0, coeff[i, 0], coeff[i, 1], color='r', alpha=0.5)
        plt.text(coeff[i, 0] * 1.15, coeff[i, 1] * 1.15, f"{labels[i]} ({coeff[i, 0]:.2f}, {coeff[i, 1]:.2f})", fontsize=12)

    plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]:.2%})")
    plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]:.2%})")
    plt.grid()

biplot_with_numbers(X_pca, pca.components_.T, feature_names)
plt.show()
```

# pca package

```python
from pca import pca


# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target
feature_names = data.feature_names

# Perform PCA with 2 components
model = pca(n_components=2)
results = model.fit_transform(X)

# Create biplot using pca package
fig, ax = model.biplot(n_feat=4, legend=False)
plt.show()
```
