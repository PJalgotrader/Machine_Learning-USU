## Evaluation metrics:

One of the downsides of sklearn package is that, for linear regression models, the only built-in evaluation metric is R-squared (score). We need to mannually construct the other evaluation metrics like MSE, RMSE, adjusted R-squared, AIC, BIC, and etc. 

Alternatively we could have used statsmodel package and get a nice summary table. 

### R-squared

```python
# The train set R-squared of the regression
reg_model.score(X_train,y_train)
```

```python
print('Training data R-squared:', np.round(reg_model.score(X_train, y_train),3))
print('Test data R-squared:', np.round(reg_model.score(X_test, y_test),3))
```

### Adjusted R-squared

```python
X_train.shape
```

```python
def adj_R2(X,y): 
    R2 = reg_model.score(X,y)
    n  = X.shape[0]
    k  = X.shape[1]
    return np.round(1-(1-R2)*(n-1)/(n-k-1),3)
```

```python
adj_R2(X_train,y_train)
```

```python
adj_R2(X_test,y_test)
```

### MSE and RMSE 
This is the most common metric used for comparing linear regression models!

```python
df_predictions.head()
```

```python
MSE_test = np.mean(np.square(df_predictions['resid']))
np.round(MSE_test,3)
```

```python
RMSE_test = np.sqrt(MSE_test)
np.round(RMSE_test,3)
```

## Cross Validation 

Estimating the test set evaluation metrics using cross validation!
Remember, we are interested in model performance in the test set not train set!! and we can estimate that number by finding the cross validated version. 

```python
from sklearn.model_selection import cross_val_score
import sklearn.metrics
```

```python
# What metrics are available for cross validation in sklearn?
sorted(sklearn.metrics.SCORERS.keys())
```

```python
my_estimator = LinearRegression()
```

```python
R2 = cross_val_score(estimator=my_estimator, X=X_train, y=y_train, cv=5, scoring="r2")
```

```python
R2
```

```python
R2_CV = np.mean(R2)
np.round(R2_CV,3)
```

This is the estimated version of $R^2_{test}$ which was about 0.694!

```python
MSE = -cross_val_score(estimator=my_estimator, X=X_train, y=y_train, cv=10, scoring="neg_mean_squared_error")
```

```python
MSE
```

```python
MSE_CV = np.mean(MSE)
np.round(MSE_CV,3)
```

This is the estimated version of **MSE_test** which was about 26.082!

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
RMSE_CV = np.mean(RMSE)
np.round(RMSE_CV,3)
```

This is the estimated version of **RMSE_test** which was about 5.107!
